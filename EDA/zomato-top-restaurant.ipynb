# *Best Rated restaurant* #

```python

df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

restaurant_ratings = df.groupby('Restaurant Name')['Rating'].mean()

top_rated_restaurants = restaurant_ratings.sort_values(ascending=False).head(10).reset_index()

print(top_rated_restaurants.head(10))
```

```python
plt.figure(figsize=(10, 6))  
sns.barplot(x='Restaurant Name', y='Rating', data=top_rated_restaurants, palette='viridis')
plt.xticks(rotation=90)  
plt.title('Top 10 Rated Restaurants')
plt.xlabel('Restaurant Name')
plt.ylabel('Mean Rating')
plt.show()
```

# *Most popular restaurant* #

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'Rating' column contains numerical values and 'Restaurant Name' column contains the restaurant names
# Replace 'df' with the name of your DataFrame

df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

restaurant_ratings = df.groupby('Restaurant Name')['Rating'].sum()

top_rated_restaurants = restaurant_ratings.sort_values(ascending=False).head(10).reset_index()

print(top_rated_restaurants.head(10))

plt.figure(figsize=(10, 6))  
sns.barplot(x='Restaurant Name', y='Rating', data=top_rated_restaurants, palette='viridis')
plt.xticks(rotation=90)  
plt.title('Top 10 Popular Restaurants by Ratings Sum')
plt.xlabel('Restaurant Name')
plt.ylabel('Total Ratings')
plt.show()
```

```python

df['Safety Measure'].value_counts().plot(kind='pie', autopct='%1.1f%%', colormap='Pastel1')

plt.axis('equal')  
plt.title('Safety Measures Distribution')

plt.show()
```

```python
def rating_to_cat(rating):
    if rating>=4:
        return '4+'
    elif 3 <= rating <4 :
        return '3+'
    elif 2<=rating<3:
        return '2+'
    else:
        return '1+'
```

```python
df['Rating_count']=df['Rating'].apply(rating_to_cat)
```

```python
df['Average Price'] = df['Average Price'].str.extract('(\d+)').fillna(0).astype(int)
```

```python
df.sample(10)
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Average Price'], y=df['Rating'], hue=df['Rating_count'], alpha=0.7)
plt.title('Relation between Average Price and Rating')
plt.xlabel('Average Price')
plt.ylabel('Rating')
plt.legend(title='Rating Category')
plt.show()
```

```python

df = df[df['Location'] != '-']
restaurant_count = df.groupby("Location")["Restaurant Name"].count().reset_index(name='count')
restaurant_count
```

```python
import plotly.express as px
from geopy.geocoders import ArcGIS


df = df[df['Location'] != '-']
restaurant_count = df.groupby("Location")["Restaurant Name"].count().reset_index(name='count')

nom = ArcGIS()
restaurant_count[["Lat", "Long"]] = restaurant_count['Location'].apply(lambda x: nom.geocode(x)[1]).apply(pd.Series)
fig = px.scatter_mapbox(
    restaurant_count,
    lon='Long',
    lat='Lat',
    zoom=3,
    color='Location',
    size='count',
    title='Restaurant Distribution Across Locations',
    mapbox_style='open-street-map',
)

fig.update_layout(margin={"r": 0, 't': 50, 'l': 0, 'b': 10})
fig.show()
```
