```python
import pandas as pd
df=pd.read_csv("/content/tested.csv")
```

```python
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Cabin','Embarked'],inplace=True,axis='columns')
```

```python
df.drop(['Ticket'],axis='columns',inplace=True)
```

```python

df
```

```python
target=df.Survived
input_var=df.drop('Survived',axis='columns')
```

```python
input_var
```

```python
dummies=pd.get_dummies(input_var.Sex)
dummies
```

```python
input_var=pd.concat([input_var,dummies],axis='columns')
```

```python
input_var
```

```python

input_var.drop('Sex',inplace=True,axis='columns')
```

```python
input_var
```

```python
input_var.columns[input_var.isna().any()]
```

```python
input_var[['Age','Fare']]
```

```python
input_var.Age=input_var.Age.fillna(input_var.Age.mean())
```

```python
input_var.Fare=input_var.Fare.fillna(input_var.Fare.mean())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(input_var , target , test_size=0.20)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```
